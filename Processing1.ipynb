```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('BollywoodMovieDetail.csv')
```

```python
data
```

```python
data.columns
```

```python
data.releaseYear.value_counts(dropna=False).sort_index().plot(kind='barh',figsize=(15,10))
plt.show()
```

```python
data = data.loc[:,['title','directors','actors','genre']]
```

```python
data
```

```python
data['directors'] = data['directors'].replace(np.nan, 'unknown')
data['actors'] = data['actors'].replace(np.nan, 'unknown')
```

```python
data['genre'] = data['genre'].str.replace('|', ' ')
data['directors'] = data['directors'].str.replace('|', ' ')
data['actors'] = data['actors'].str.replace('|', ' ')
```

```python
data
```

```python
data['title'] = data['title'].str.lower()
```

```python
data['title'][6]
```

```python
data['comb'] = data['directors'] + ' ' + data['actors'] + ' '+ data['genre']
```

```python
data
```

```python
data.to_csv('data.csv',index=False)
```
